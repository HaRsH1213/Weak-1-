```python
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title("Distribution of Generated Power(KW)")
plt.ylabel('Frequency')
plt.xlabel('Generated Power (KW)')
plt.show()
```

## EDA

```python
df[df.columns[:9]].hist(bins=30,figsize=(15,10))
plt.show
```

```python
plt.figure(figsize=(15,30))
for i, columns in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[columns],df['generated_power_kw'])
    plt.title(f'{columns} vs Genrated Power (KW)')
    plt.xlabel(columns)
    plt.ylabel('Genrated Power (KW)')
plt.tight_layout()
plt.show
```
